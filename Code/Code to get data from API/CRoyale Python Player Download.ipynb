```python

# Read the CSV file into a DataFrame
df2 = pd.read_csv('class_royale_battlelog_dataset.csv')

# Extract the 'player_tag' and 'opponent_tag' columns and combine them
tags = pd.concat([df2['player_tag'], df2['opponent_tag']])

# Remove duplicates from the combined series
unique_tags = tags.drop_duplicates().tolist()

# Now unique_tags contains all unique player and opponent tags without duplicates
print(unique_tags)
```

```python
print(len(unique_tags))
```

```python
players_data = []

# Loop through the unique_tags in slices of 200
for i in range(0, len(unique_tags), 200):
    # Process the current slice
    for tag in unique_tags[i:i+200]:
        data = fetch_data(tag)
        if data:  # If data is not None
            # Extract the required fields and append to the list
            player_info = {
                "tag": tag,
                "name": data.get("name"),
                "expLevel": data.get("expLevel"),
                "wins": data.get("wins"),
                "losses": data.get("losses"),
                "battleCount": data.get("battleCount"),
                "threeCrownWins": data.get("threeCrownWins"),
                "totalExpPoints": data.get("totalExpPoints")
            }
            players_data.append(player_info)
    
    # Convert the current slice of data to a DataFrame
    slice_df = pd.DataFrame(players_data)
    # Append the current slice DataFrame to the CSV file
    if i == 0:
        # If it's the first slice, write a new file with headers
        slice_df.to_csv('class_royale_player_dataset.csv', mode='w', header=True, index=False)
    else:
        # Otherwise, append without headers
        slice_df.to_csv('class_royale_player_dataset.csv', mode='a', header=False, index=False)
    
    # Clear players_data list for the next batch
    players_data.clear()

print("All data saved to class_royale_player_dataset.csv")
```
