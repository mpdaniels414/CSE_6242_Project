```python
import pandas as pd

# Merge player stats onto battles using player_tag
results_ml = battles.merge(player_stats.add_prefix('player_'),
                           left_on='player_tag',
                           right_on='player_tag',
                           how='left')

# Merge opponent stats onto battles using opponent_tag
results_ml = results_ml.merge(player_stats.add_prefix('opponent_'),
                              left_on='opponent_tag',
                              right_on='opponent_tag',
                              how='left')

# results_ml now contains the merged data
```

```python
# make sure all columns contain numeric values only
import pandas as pd

# Assuming 'data' is your DataFrame
string_columns = results_ml.select_dtypes(include=['object']).columns

# Display string columns
print(f'Columns with string values: {string_columns}')

# Convert string columns to numerical values where possible
for col in string_columns:
    try:
        results_ml[col] = pd.to_numeric(results_ml[col], errors='coerce')
    except Exception as e:
        print(f'Could not convert column {col} to numeric: {e}')
```

```python
# This will fill NaN values with the mean of each column.
for column in results_ml.columns:
    results_ml[column].fillna(results_ml[column].mean(), inplace=True)
```
